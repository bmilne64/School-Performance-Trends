```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_path = "Resources/schools_complete.csv"
student_path = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_path)
student_data = pd.read_csv(student_path)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```





## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = school_data_complete['school_name'].nunique()

#Calculate the total number of students                                 
total_students = school_data_complete['student_name'].count()

#Calculate the total budget
school_data_new = school_data_complete.drop_duplicates(subset=['budget'])
total_budget = school_data_new["budget"].sum()

#Calculate the average math score                                  
average_math = school_data_complete["math_score"].mean()

#Calculate the average reading score
average_reading = school_data_complete["reading_score"].mean()

#Calculate the percentage of students with a passing math score (70 or greater)                                       
passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70,"math_score"].count()

math_percent = (passing_math /(school_data_complete['math_score'].count()) * 100)

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70, "reading_score"].count()

reading_percent = (passing_reading /(school_data_complete['reading_score'].count()) * 100)

#Calculate the percentage of students who passed math and reading (% Overall Passing)                                           
passing_both = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70), "reading_score"].count()

overall_percent = (passing_both/(school_data_complete['reading_score'].count()) * 100)

#Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total Schools" : [total_schools], 
                                 "Total Students" : [total_students], 
                                 "Total Budget" : [total_budget],
                                 "Average Math Score": [average_math],
                                 "Average Reading Score": [average_reading],
                                 '% Passing Math' : [math_percent],
                                 '% Passing Reading': [reading_percent], 
                                 '% Overall Passing': [overall_percent]})

#format columns

district_summary["Total Budget"] = (district_summary["Total Budget"]).map("${:,.2f}".format)

district_summary["Total Students"] = (district_summary["Total Students"]).map("{:,.0f}".format)

#print new dataframe
district_summary.head()
```

## School Summary






* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  



* Create a dataframe to hold the above results

```python
#Group table by school name
grouped_schools = school_data_complete.groupby("school_name")

#School Type
school_type = grouped_schools['type'].unique()

#Total Students
student_count = grouped_schools["student_name"].count()

#Total School Budget
school_budget = grouped_schools["budget"].mean()

#calculate per student budget
per_student_budget = school_budget/student_count

# Average Math Score
average_math_score = grouped_schools["math_score"].mean()

#Average Reading Score
average_reading_score = grouped_schools["reading_score"].mean()

# % Passing Math
passing_math2 = school_data_complete[school_data_complete['math_score'] >= 70].groupby("school_name")['math_score'].count()

m_percent = (passing_math2/student_count)*100
```

```python
# % Passing Reading
passing_reading2 = school_data_complete[school_data_complete['reading_score'] >= 70].groupby("school_name")['reading_score'].count()

r_percent = (passing_reading2/student_count)*100

# % Overall Passing (The percentage of students that passed math and reading.)
passing_both2 = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby("school_name")['math_score'].count()

both_percent = (passing_both2/student_count)*100

#Create a dataframe to hold the above results
grouped_schools_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": student_count, 
    "Total School Budget": school_budget,
    'Per Student Budget': per_student_budget, 
    "Average Math Score": average_math_score,
    'Average Reading Score': average_reading_score, 
    "% Passing Math": m_percent,
    '% Passing Reading': r_percent,
    '% Overall Passing': both_percent,
})

#format columns

grouped_schools_df["Total School Budget"] = (grouped_schools_df["Total School Budget"]).map("${:,.2f}".format)

grouped_schools_df["Total Students"] = (grouped_schools_df["Total Students"]).map("{:,.0f}".format)

grouped_schools_df["School Type"] = grouped_schools_df["School Type"].apply(lambda row: row[0])

#print dataframe
grouped_schools_df
```

## Scores by School Spending


* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#merge datasets to include per student budget
school_data_spending = pd.merge(school_data_complete, grouped_schools_df, how="inner", on=["school_name", "school_name"])

#create bins
spending_bins = [0, 585, 630, 645, 680]

#create labels 
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

#add a column for Spending Ranges (Per Student) to existing dataframe
school_data_spending["Spending Ranges (Per Student)"] = pd.cut(school_data_spending["Per Student Budget"], bins=spending_bins, labels=labels)
school_data_spending
type1 = school_data_spending.groupby("school_name").mean()
type1["Spending Ranges (Per Student)"] = pd.cut(type1["Per Student Budget"], bins=spending_bins, labels=labels)
type1
```

```python
#group by Spending Ranges (Per Student)
schools_spending = school_data_spending.groupby("Spending Ranges (Per Student)")

#find average math score
spending_average_math = schools_spending["math_score"].mean()

#find average reading score
spending_average_reading = schools_spending["reading_score"].mean()

#count of students over 70% in math
spending_math_percent = school_data_spending[school_data_spending['math_score'] >= 70].groupby("Spending Ranges (Per Student)")['math_score'].count()

#find student count 
student_count = schools_spending["student_name"].count()

#find percent of students over 70% in math
spending_math = (spending_math_percent/student_count)*100

#count of students over 70% in reading
spending_reading_percent = school_data_spending[school_data_spending['reading_score'] >= 70].groupby("Spending Ranges (Per Student)")['reading_score'].count()

#find percent of students over 70% in reading
spending_reading = (spending_reading_percent/student_count)*100

#count of students over 70% in math and reading
spending_overall_percent = school_data_spending[(school_data_spending['reading_score'] >= 70) & (school_data_spending['math_score'] >= 70)].groupby("Spending Ranges (Per Student)")['reading_score'].count()

#find percent of students over 70% in math and reading
spending_overall = (spending_overall_percent/student_count)*100

#create new dataframe
scores_by_school_spending = pd.DataFrame({"Average Math Score": spending_average_math,
                                      'Average Reading Score':spending_average_reading,
                                      "% Passing Math": spending_math,
                                      "% Passing Reading":spending_reading,
                                      "% Overall Passing": spending_overall})

# #format columns
scores_by_school_spending["% Passing Math"] = (scores_by_school_spending["% Passing Math"]).map("{:.2f}%".format)

scores_by_school_spending["% Passing Reading"] = (scores_by_school_spending["% Passing Reading"]).map("{:.2f}%".format)

scores_by_school_spending["% Overall Passing"] = (scores_by_school_spending["% Overall Passing"]).map("{:.2f}%".format)

scores_by_school_spending["Average Math Score"] = (scores_by_school_spending["Average Math Score"]).map("{:,.2f}".format)

scores_by_school_spending["Average Reading Score"] = (scores_by_school_spending["Average Reading Score"]).map("{:,.2f}".format)

#print dataframe
scores_by_school_spending.head()
```
